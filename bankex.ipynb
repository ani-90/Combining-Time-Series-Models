```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
au=pd.Series(pd.read_csv('/content/au.csv')['Close Price'],name='AU')
axis=pd.Series(pd.read_csv('/content/axis.csv')['Close Price'],name='Axis')
bob=pd.Series(pd.read_csv('/content/bob.csv')['Close Price'],name='BoB')
fed=pd.Series(pd.read_csv('/content/federal.csv')['Close Price'],name='FB')
hdfc=pd.Series(pd.read_csv('/content/hdfc.csv')['Close Price'],name='HDFC')
icici=pd.Series(pd.read_csv('/content/icici.csv')['Close Price'],name='ICICI')
idfc=pd.Series(pd.read_csv('/content/idfc.csv')['Close Price'],name='IDFC')
indusind=pd.Series(pd.read_csv('/content/indusind.csv')['Close Price'],name='INDUSIND')
kotak=pd.Series(pd.read_csv('/content/kotak.csv')['Close Price'],name='Kotak')
sbi=pd.Series(pd.read_csv('/content/sbin.csv')['Close Price'],name='SBI')
```

```python
df=pd.concat([au,axis,bob,fed,hdfc,icici,idfc,indusind,kotak,sbi],axis=1)
df.head()
```

```python
## plotting correlation between the closing stock prices of the above banks
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='viridis')
plt.show()
```

```python
corr=df.corr()
distance=1-corr.pow(2)
distance
```

```python
## classifying relationships based on the values of distance matrix

weak=[]
moderate=[]
strong=[]
cols=list(distance.columns)
for i in range(len(cols)-1):
  c1=cols[i]
  for j in range(i+1,len(cols)):
    c2=cols[j]
    val=distance[c1][c2]

    if 0.75<=val<=1:
      if [c1,c2] and [c2,c1] not in weak:
        weak.append([c1,c2])
    elif 0.36<=val<0.75:
      if [c1,c2] and [c2,c1] not in moderate:
        moderate.append([c1,c2])

    elif 0<=val<=0.36:
      if [c1,c2] and [c2,c1] not in strong:
        strong.append([c1,c2])

```

```python
print('Banks with weak relationship: ',weak)
print('Banks with moderate relationship: ',moderate)
print('Banks with Strong relationship: ',strong)
```

```python
## Applying Principal Component Analysis for the distance matrix

from sklearn.decomposition import PCA
pca=PCA(n_components=4)
pca.fit(distance)
```

```python
## transforming the data

pca_data=pd.DataFrame(pca.fit_transform(distance))
pca_data

for i in range(len(distance.columns)):
  print(distance.columns[i],np.linalg.norm(pca_data.iloc[0:1,:]))
```

```python

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Plot the explained variance
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
linkage_matrix = linkage(pca_data, method='complete', metric='euclidean')

plt.figure(figsize=(10, 8))
dendrogram(linkage_matrix, labels=distance.index, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (PCA Data)')
plt.xlabel('Banks')
plt.ylabel('Distance')
plt.show()

## 2,4,5,6
```
