```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
au=pd.Series(pd.read_csv('/content/au.csv')['Close Price'],name='AU')
axis=pd.Series(pd.read_csv('/content/axis.csv')['Close Price'],name='Axis')
bob=pd.Series(pd.read_csv('/content/bob.csv')['Close Price'],name='BoB')
fed=pd.Series(pd.read_csv('/content/federal.csv')['Close Price'],name='FB')
hdfc=pd.Series(pd.read_csv('/content/hdfc.csv')['Close Price'],name='HDFC')
icici=pd.Series(pd.read_csv('/content/icici.csv')['Close Price'],name='ICICI')
idfc=pd.Series(pd.read_csv('/content/idfc.csv')['Close Price'],name='IDFC')
indusind=pd.Series(pd.read_csv('/content/indusind.csv')['Close Price'],name='INDUSIND')
kotak=pd.Series(pd.read_csv('/content/kotak.csv')['Close Price'],name='Kotak')
sbi=pd.Series(pd.read_csv('/content/sbin.csv')['Close Price'],name='SBI')
```

```python
df=pd.concat([au,axis,bob,fed,hdfc,icici,idfc,indusind,kotak,sbi],axis=1)
df.head()
```

```python
## plotting correlation between the closing stock prices of the above banks
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='viridis')
plt.show()
```

```python
corr=df.corr()
distance=1-corr.pow(2)
distance
```

```python
## classifying relationships based on the values of distance matrix

weak=[]
moderate=[]
strong=[]
cols=list(distance.columns)
for i in range(len(cols)-1):
  c1=cols[i]
  for j in range(i+1,len(cols)):
    c2=cols[j]
    val=distance[c1][c2]

    if 0.75<=val<=1:
      if [c1,c2] and [c2,c1] not in weak:
        weak.append([c1,c2])
    elif 0.36<=val<0.75:
      if [c1,c2] and [c2,c1] not in moderate:
        moderate.append([c1,c2])

    elif 0<=val<=0.36:
      if [c1,c2] and [c2,c1] not in strong:
        strong.append([c1,c2])

```

```python
print('Banks with weak relationship: ',weak)
print('Banks with moderate relationship: ',moderate)
print('Banks with Strong relationship: ',strong)
```

```python
## Applying Principal Component Analysis for the distance matrix

from sklearn.decomposition import PCA
pca=PCA(n_components=4)
pca.fit(distance)
```

```python
## transforming the data

pca_data=pd.DataFrame(pca.fit_transform(distance))
```

```python

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Plot the explained variance
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python

import pandas as pd
from sklearn.metrics import pairwise_distances
from scipy.sparse.csgraph import minimum_spanning_tree

pca_data
dist = pairwise_distances(pca_data)

mst = minimum_spanning_tree(dist)

threshold1=0.1
threshold2=0.2
threshold3=0.25
threshold4=0.65

clusters1 = []
clusters2 = []
clusters3 = []
clusters4 = []


for i in range(len(mst.indices)):
    for j in range(i + 1, len(mst.indices)):
        if mst[i, j] != 0 and mst[i, j] <= threshold1:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters1:
                if i in cluster or j in cluster:
                  cluster.add(bank_index[i])
                  cluster.add(bank_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters1.append([bank_index[i], bank_index[j],round(mst[i,j],3)])

        if mst[i, j] != 0 and mst[i, j] <= threshold2:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters2:
                if i in cluster or j in cluster:
                  cluster.add(bank_index[i])
                  cluster.add(bank_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters2.append([bank_index[i], bank_index[j],round(mst[i,j],3)])
        if mst[i, j] != 0 and mst[i, j] <= threshold3:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters3:
                if i in cluster or j in cluster:
                  cluster.add(bank_index[i])
                  cluster.add(bank_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters3.append([bank_index[i], bank_index[j],round(mst[i,j],3)])
        if mst[i, j] != 0 and mst[i, j] <= threshold4:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters4:
                if i in cluster or j in cluster:
                  cluster.add(bank_index[i])
                  cluster.add(bank_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters4.append([bank_index[i], bank_index[j],round(mst[i,j],3)])

print("Clusters for 0.1:", clusters1)
print("Clusters for 0.25:", clusters2)
print("Clusters for 0.3:", clusters3)
print("Clusters for 0.65:", clusters4)

# print(mst.toarray())
```

```python
import networkx as nx
G=nx.Graph()
G.add_weighted_edges_from(clusters3)
T=nx.minimum_spanning_tree(G)
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,font_weight='bold',node_size=1000,node_color='green')



nx.draw_networkx_edges(T,pos,edge_color='red',width=1)

labels=nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.title('Cluster for 0.3')
plt.show()
```
