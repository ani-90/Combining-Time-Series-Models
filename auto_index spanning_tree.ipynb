```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apollo=pd.Series(pd.read_csv('/content/apollo tyres.csv')['Close Price'],name='apollo')
ashokleyland=pd.Series(pd.read_csv('/content/ashokleyland.csv')['Close Price'],name='ashokleyland')
bajaj=pd.Series(pd.read_csv('/content/bajaj.csv')['Close Price'],name='bajaj')
balkrishna_motors=pd.Series(pd.read_csv('/content/balkrishna industries.csv')['Close Price'],name='balkrishna_motors')
bosch=pd.Series(pd.read_csv('/content/bosch.csv')['Close Price'],name='bosch')
cummins=pd.Series(pd.read_csv('/content/cummins.csv')['Close Price'],name='cummins')
eichermotors=pd.Series(pd.read_csv('/content/eichermotors.csv')['Close Price'],name='eichermotors')
heromotors=pd.Series(pd.read_csv('/content/heromotors.csv')['Close Price'],name='heromotors')
mahindra=pd.Series(pd.read_csv('/content/mahindra and manindra.csv')['Close Price'],name='mahindra')
maruti=pd.Series(pd.read_csv('/content/maruti suzuki.csv')['Close Price'],name='maruti')
motherson=pd.Series(pd.read_csv('/content/motherson.csv')['Close Price'],name='motherson')
mrf=pd.Series(pd.read_csv('/content/mrf.csv')['Close Price'],name='mrf')
tatamotors=pd.Series(pd.read_csv('/content/tatamotors.csv')['Close Price'],name='tatamotors')
tube_investments=pd.Series(pd.read_csv('/content/tube investments.csv')['Close Price'],name='tube investments')
tvs_motors=pd.Series(pd.read_csv('/content/tvs motors.csv')['Close Price'],name='tvs motors')
```

```python
df=pd.concat([apollo,ashokleyland,bajaj,balkrishna_motors,bosch,cummins,eichermotors,heromotors,mahindra,maruti,motherson,mrf,tatamotors,tube_investments,tvs_motors],axis=1)
df.head()
```

```python
## plotting correlation between the closing stock prices of the above automobile companies
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='viridis')
plt.show()
```

```python
## distance values based on the correlation

corr=df.corr()
distance=1-corr.pow(2)
distance
```

```python
## classifying relationships based on the values of distance matrix

weak=[]
moderate=[]
strong=[]
cols=list(distance.columns)
for i in range(len(cols)-1):
  c1=cols[i]
  for j in range(i+1,len(cols)):
    c2=cols[j]
    val=distance[c1][c2]

    if 0.75<=val<=1:
      if [c1,c2] and [c2,c1] not in weak:
        weak.append([c1,c2])
    elif 0.36<=val<0.75:
      if [c1,c2] and [c2,c1] not in moderate:
        moderate.append([c1,c2])

    elif 0<=val<=0.36:
      if [c1,c2] and [c2,c1] not in strong:
        strong.append([c1,c2])

```

```python
print('auto mobile companies with weak relationship: ',weak)
print('auto mobile companies with moderate relationship: ',moderate)
print('auto mobile companies with Strong relationship: ',strong)
```

```python
## Applying Principal Component Analysis for the distance matrix

from sklearn.decomposition import PCA
pca=PCA(n_components=4)
pca.fit(distance)
```

```python
pca_data=pd.DataFrame(pca.fit_transform(distance))
pca_data
```

```python

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Plot the explained variance
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
auto_index={
    0:'apollo',
    1:'ashok leyland',
    2:'bajaj',
    3:'balkrishna industries',
    4:'bosch',
    5:'cummins',
    6:'eichermotors',
    7:'heromotors',
    8:'manindra',
    9:'maruti suzuki',
    10:'motherson',
    11:'mrf',
    12:'tatamotors',
    13:'tube investments',
    14:'tvs motors'
}
```

```python

import pandas as pd
from sklearn.metrics import pairwise_distances
from scipy.sparse.csgraph import minimum_spanning_tree

pca_data
dist = pairwise_distances(pca_data)

mst = minimum_spanning_tree(dist)

threshold1=0.1
threshold2=0.4
threshold3=0.7
threshold4=1.5

clusters1 = []
clusters2 = []
clusters3 = []
clusters4 = []


for i in range(len(mst.indices)):
    for j in range(i + 1, len(mst.indices)):
        if mst[i, j] != 0 and mst[i, j] <= threshold1:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters1:
                if i in cluster or j in cluster:
                  cluster.add(auto_index[i])
                  cluster.add(auto_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters1.append([auto_index[i], auto_index[j],round(mst[i,j],3)])

        if mst[i, j] != 0 and mst[i, j] <= threshold2:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters2:
                if i in cluster or j in cluster:
                  cluster.add(auto_index[i])
                  cluster.add(auto_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters2.append([auto_index[i], auto_index[j],round(mst[i,j],3)])
        if mst[i, j] != 0 and mst[i, j] <= threshold3:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters3:
                if i in cluster or j in cluster:
                  cluster.add(auto_index[i])
                  cluster.add(auto_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters3.append([auto_index[i], auto_index[j],round(mst[i,j],3)])
        if mst[i, j] != 0 and mst[i, j] <= threshold4:
            # Check if the vertices are already in clusters
            found = False
            for cluster in clusters4:
                if i in cluster or j in cluster:
                  cluster.add(auto_index[i])
                  cluster.add(auto_index[j])
                  cluster.add(round(mst[i,j],3))
                  found = True
                  break
            if not found:
              clusters4.append([auto_index[i], auto_index[j],round(mst[i,j],3)])

print("Clusters for 0.1:", clusters1)
print("Clusters for 0.4:", clusters2)
print("Clusters for 0.7:", clusters3)
print("Clusters for 1.5:", clusters4)

```
